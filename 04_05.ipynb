```python
#Escrever um programa que o usuario digta um numero e o programa imprime em ordem decrescente ate zero

numero = int (input("Digite um número: "))

for i in range (numero, -1, -1):
  print (i)
```

```python
# Peça uma palavra e conte as letras: Ex: Phyton - A palavra tem 06 letras

palavra = input("Digite uma palavra: ")

quantidade = len(palavra)
print (f"A palavra {palavra} tem {quantidade} de letras")
```

```python
# Peça uma palavra e conte as vogais

palavra = input("Digite uma palavra: ")
vogais = "a e i o u á à â ã é ê í ó ô ú ç"
contador = 0

for i in palavra.lower():
    if i in vogais:
        contador = contador + 1
print(f"A palavra '{palavra}' tem {contador} vogais.")
```
