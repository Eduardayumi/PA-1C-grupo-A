Escrever um algoritímo e um programa que recebe 04 notas do aluno e calcula a média depois mostrar a média

```python
# Entrada

nota_1 = 10
nota_2 = 5
nota_3 = 8
nota_4 = 7

# processamento

Média = (nota_1 + nota_2 + nota_3 +nota_4) / 4

#Saída

print (f"Média: {Média}")
```
