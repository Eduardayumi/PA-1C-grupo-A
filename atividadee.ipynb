```python
#Questão 2 – Cálculo de desconto

#Crie um programa que leia o valor de uma compra e aplique o desconto conforme as regras:
#	•	Até R$100: sem desconto
#	•	De R$101 a R$500: 10% de desconto
#	•	Acima de R$500: 20% de desconto
#Ao final, mostre o valor final da compra.


valor = int (input ("Digite o valor da compra: "))

if  valor <= 100: 
 print (f"O valor final é R$ {valor}") 
elif valor > 100 and valor <= 500: 
 print (f"O valor final é R$ {valor - valor * 0.10 } com 10% de descounto") 
else:
 print (f"O valor é final é R$ {valor - valor * 0.20 } com 20% de descounto") 
```

```python
#Questão 4 – Sistema de energia (jogo)

#Crie um programa que leia a energia de um personagem (de 0 a 100) e classifique:
#	•	0: Morto
#	•	1 a 30: Crítico
#	•	31 a 70: Normal
#	•	71 a 100: Saudável

energia = float (input ("Digite o número da energia do personagem (0 a 100): "))

if energia <= 30:
 print ("energia crítica")
elif energia == 0:
 print ("Morto")
elif energia <= 70:
 print ("Normal")
elif energia <= 100:
 print("Saúdavel")
else:
 print ("Valor inválido, digite novamente")
```

```python
# Questão 5 – Média escolar

#Crie um programa que leia a média final de um aluno e classifique:
# 	•	Média maior ou igual a 7: Aprovado
#	•	Média entre 5 e 6.9: Recuperação
#	•	Média menor que 5: Reprovado

a = float (input ("Digite o valor da nota: "))
b = float (input ("Digite o valor da nota: "))
c = float (input ("Digite o valor da nota: "))

média = ( a + b + c ) / 3

if média >= 7:
 print (f"Sua média foi {média}, Aprovado ")
elif média > 5:
 print (f"Sua média foi {média}, está em recuperação ")
else:
 print ("Reprovado")

```
