```python
# 10 = Pirâmide númerica simetrica

#Faça um programa que receba um número N = Imprima uma pirâmide perfeitamente centralizada
#Exemplo N = 5:

def piramide_simetrica(n):
    for i in range(1, n + 1):
        espacos = ' ' * (n - i)

        # Parte crescente
        crescente = ''.join(str(j) for j in range(1, i + 1))

        # Parte decrescente
        decrescente = ''.join(str(j) for j in range(i - 1, 0, -1))


        print(espacos + crescente + decrescente)

n = 5
piramide_simetrica(n)

```
