```python
# 10 - 1: Crie um jogo de adivinhação, o usuário tenta adivinhar um número secreto de 1 até 10.
# 2: Informar se o número secreto é maior ou menor
# 3: Se o usuaro acertar, exibir uma mensagem de "Você acertou! "
# 4: O usuário terá 3 tentativas

import random

numero_secreto = random.randint(1, 10)
limite_tentativas = 3

print("Tente adivinhar o número secreto de 1 a 10!")
print("Você tem 3 tentativas.")

for tentativa in range(1, limite_tentativas + 1):
    palpite = int(input(f"\nTentativa {tentativa} de {limite_tentativas}. Qual é o seu palpite? "))

    if palpite == numero_secreto:
        print("Você acertou!")
        break
    elif palpite < numero_secreto:
        print("O número secreto é maior.")
    else:
        print("O número secreto é menor.")

if palpite != numero_secreto:
    print(f"\nSuas tentativas acabaram! O número secreto era {numero_secreto}.")
```
