Fazer um programa que converte a temperatura de celsius para Fahrenheit.

Entrada : graus celsius

Processamento: Converter a temperatura para Fahrenheit

Saída : Mostrar a temperatura em Fahrenheit

F = (C x 9/5) + 32

```python
# Entrada

Graus_celsius = float (input ("digite os graus celsius: "))

# processamento

Fahrenheit = (Graus_celsius * 9/5) + 32

# Saída

print (f"{Graus_celsius} graus em  Fahrenheit é {Fahrenheit}")
```
