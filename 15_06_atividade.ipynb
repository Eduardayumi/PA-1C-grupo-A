```python
#---------------------------------------------------------
#QUESTÃO 02 – Soma dos Números (2 pontos)
#---------------------------------------------------------
#Solicite ao usuário um número inteiro positivo.
#Utilizando um laço de repetição, calcule e exiba a soma de todos os números de 1 até o número informado.


numero = int(input("Digite um número inteiro positivo: "))

soma = 0

for i in range(1, numero + 1):
    soma += i
passos = [str(x) for x in range(numero, 0, -1)]

print (f"Resultado: {soma}")

print(f"{' + '.join(passos)} = {soma} ")
```

```python
#---------------------------------------------------------
#QUESTÃO 05 – Caixa Eletrônico (5 pontos)
#---------------------------------------------------------

#Um caixa eletrônico possui notas de: R$ 100 R$ 50 R$ 20 R$ 10 R$ 5 R$ 2 R$ 1

#Solicite ao usuário o valor do saque e informe a menor quantidade possível de notas para realizar o pagamento.


valor = int(input("Digite o valor do saque: "))
nota = 100

while valor > 0:
    quantidade = valor // nota

    if quantidade > 0:
        print(f"{quantidade} nota(s) de R${nota}")
        valor %= nota

    if nota == 100:
        nota = 50
    elif nota == 50:
        nota = 20
    elif nota == 20:
        nota = 10
    elif nota == 10:
        nota = 5
    elif nota == 5:
        nota = 2
    elif nota == 2:
        nota = 1
    elif nota == 1:
        break
```
