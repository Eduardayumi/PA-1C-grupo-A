```python
# Escrever um programa que o aluno digita 4 notas dos bimestres e o programa calcula a media e mostra a sáida se o aluno foi aprovado, recuperação ou aprovado

nota_1 = float (input("Digite a nota: "))
nota_2 = float (input("Digite a nota: "))
nota_3 = float (input("Digite a nota: "))
nota_4 = float (input("Digite a nota: "))

media = (nota_1 + nota_2 + nota_3 + nota_4) / 4

if media >= 7:

  print (f"Você foi aprovado, sua média foi: {media} ")

elif media >= 5:

  print (f"Você está de recuperação, sua média foi: {media}")

else:

  print (f"Você foi reprovado, sua média foi: {media}")
```
